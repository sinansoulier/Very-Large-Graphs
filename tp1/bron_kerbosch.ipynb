```python
# import all our functions
from naive import *
from bron_kerbosch import *
import matplotlib.pyplot as plt
```

```python
# creation d'un graphe
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8])
G.add_edges_from([(1, 2), (1, 3), (1, 4), (2, 3), (2, 5), (3, 4), (3, 5), (4, 5), (4, 6), (5, 6), (5, 7), (5, 8), (6, 7), (6, 8), (7, 8)])

# another graph
G2 = nx.Graph()
G2.add_nodes_from([1, 2, 3, 4, 5])
G2.add_edges_from([(1, 2), (1, 3), (1, 4), (2, 3), (3, 4), (3, 5), (4, 5)])

# a fully connected graph
G3 = nx.Graph()
G3.add_nodes_from([1, 2, 3, 4, 5])
G3.add_edges_from([(1, 2), (1, 3), (1, 4), (1, 5), (2, 3), (2, 5), (3, 4), (3, 5), (4, 5), (4, 2)])


plt.subplot(131)
nx.draw(G, with_labels=True)
plt.subplot(132)
nx.draw(G2, with_labels=True)
plt.subplot(133)
nx.draw(G3, with_labels=True)
plt.show()
```

### Tests est_clique

```python
print("Les sommets 5, 6, 7, 8 forment une clique:",est_clique(G, [5, 6, 7, 8]))
print("Les sommets 1,2,3 forment une clique:",est_clique(G, [1, 2, 3]))
print("Les sommets 2,3,4,5 forment une clique:",est_clique(G, [2,3,4,5]))
```
